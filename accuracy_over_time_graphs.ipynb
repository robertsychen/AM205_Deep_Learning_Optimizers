```python
import numpy as np
import cPickle as pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diff_algorithms = ['ConjugateGradient', 'HessianFree', 'LBFGS', 'CustomGradientDescent', 'CustomAdam']
diff_noises = ['none', 'test', 'both']

for algo in diff_algorithms:
    plt.figure()
    for noise in diff_noises:
        data = pickle.load(open(algo + str(500) + noise + str(0) + ".pkl", "rb"))
        plt.plot(range(1,len(data)+1), data, label="Noise: " + noise)
        plt.title("Accuracy over Time: " + algo)
        plt.xlabel("Training Iteration No.")
        plt.ylabel("Accuracy (%)")
        plt.legend(loc='lower right')
```

```python
plt.figure()
for algo in diff_algorithms:
    data = pickle.load(open(algo + str(500) + "none" + str(0) + ".pkl", "rb"))
    plt.plot(range(1,len(data)+1), data, label=algo)
    plt.title("Accuracy over Time (No Noise Added)")
    plt.xlabel("Training Iteration No.")
    plt.ylabel("Accuracy (%)")
    plt.legend(loc='lower right')
```
